```python
# plotting histogram for age distribution
```

```python
plt.hist(df['age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Age_counts')
plt.title('Age Distribution')
plt.show()
```

```python
# bar graph for sex distribution
```

```python
plt.figure(figsize=(8, 6))
df['sex'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Sex')
plt.xlabel('Sex')
plt.ylabel('Sex_counts')
plt.show()
```

```python
# bar graph for smoking status distribution
```

```python
plt.figure(figsize=(8, 6))
df['smoking'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Distribution of Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.show()
```

```python
# bar graph for diabetes status distribution
```

```python
plt.figure(figsize=(8, 6))
df['diabetes'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Distribution of Diabetes Status')
plt.xlabel('Diabetes Status')
plt.ylabel('Count')
plt.show()
```
